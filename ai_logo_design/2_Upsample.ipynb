# Upsampling

I found this tutorial to be excellent: https://learnopencv.com/super-resolution-in-opencv/ For our purposes, one image upsample, we can use EDSR_x4

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
%matplotlib inline


img = cv.imread("media/panda logo.png")
```

```python
sr = cv.dnn_superres.DnnSuperResImpl_create()

path = "media/EDSR_x4.pb"

sr.readModel(path)

sr.setModel("edsr",4)

result = sr.upsample(img)

# Resized image
resized = cv.resize(img,dsize=None,fx=4,fy=4)

plt.figure(figsize=(26,18))
plt.subplot(1,3,1)
# Original image
plt.imshow(img[:,:,::-1])
plt.subplot(1,3,2)
# SR upscaled
plt.imshow(result[:,:,::-1])
plt.subplot(1,3,3)
# OpenCV upscaled
plt.imshow(resized[:,:,::-1])
plt.show()
```

```python
cv.imwrite("media/panda logo upscaled.png", result)
```
